### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)

# Store latitude and longitude in locations
locations = weather_df[["Lat", "Lng"]]
weight = weather_df["Humidity"]
```

```python
# Heatmap
figure = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=weight, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add and display
figure.add_layer(heat_layer)
figure
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Max temperature lower than 80 degrees but higher than 70. 
# Wind speed less than 10 mph. 
# Zero cloudiness.

format_cities = weather_df.dropna()

weather_fitted = format_cities.loc[(format_cities['Cloudiness'] > 0) & (format_cities['Wind Speed'] < 10)] 
weather_fitted_df = weather_fitted.loc[(weather_fitted['Max Temp'] > 70) & (weather_fitted['Max Temp'] <= 80)]
weather_fitted_df
```
